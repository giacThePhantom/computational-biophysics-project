```python
import numpy as np
import matplotlib.pyplot as plt

def plot_xvg(filename, title, xlab, ylab, datalab, color='red'):
    x, y = [], []
    with open(filename) as f:
        for line in f:
            if line[0] != '#' and line[0] != '@':
                cols = line.split()
                if len(cols) == 2:
                    x.append(float(cols[0]))
                    y.append(float(cols[1]))
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.set_title(title)    
    ax1.set_xlabel(xlab)
    ax1.set_ylabel(ylab)
    ax1.plot(x,y, c=color, label=datalab)
    leg = ax1.legend()
    plt.show()  
```

```python
plot_xvg('../../data/em_output/potential.xvg', "Potential during energy minimization", "Time", "Potential (kJ/Mol)", "Potential energy", color = 'blue')
```

```python
plot_xvg('../../data/nvt_output/temperature.xvg', "Temperature during NVT", "Time", "Temperature (K)", "Temperature", color = 'red')
```

```python
plot_xvg('../../data/npt_output/pressure.xvg', "Pressure during NPT", "Time", "Pressure (bar)", "Pressure", color='green')
```
